```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

```python
driver=webdriver.Chrome()
driver.get("https://www.naukri.com/")
```

```python
#Entering the designation and Location-- using xpath

designation=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[1]/div/div/div/div[1]/div/input')
designation.send_keys('Data Analyst')

#Location

location=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys('Bangalore')


#For Searching

search=driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[6]')
search.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#Lets scrap the Job title

title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
    
#Scrapping Location
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for b in location_tags[0:10]:
    location=b.text
    job_location.append(location)
    
    
#Scrapping Company Name
Company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for a in Company_tags[0:10]:
    company=a.text
    company_name.append(company)
    
#Scrapping Experience
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
#Crosscheck the length it should be same
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
#lets prepare the table using DataFrame
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company_name':company_name,'Experience':experience_required})
df
```

```python
#Question8 Solution---Authors,Quote and Quote_Name

driver = webdriver.Chrome()
driver.get('http://www.azquotes.com/')
```

```python
#lets Scrap the data 
Quote=[]
Author=[]
Type_Quote=[]
```

```python
#Top 1000 quotes
for page in range(10):
    quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for a in quote:
        Quote.append(a.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
#Top 1000 Authors

for b in range(10):
    author=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for c in author:
        Author.append(c.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
#Top 1000 Type of Quote

for d in range(10):
    Type=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for e in Type:
        Type_Quote.append(e.text)
    next_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_button.click()
    time.sleep(3)
```

```python
#Question5 Solution--Iphone review, rating and Full review

driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market")
```

```python
Rating=[]
Review=[]
Fullreview=[]
```

```python
#lets extract the rating

for Page1 in range(10):
    rate=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for a1 in rate:
        Rating.append(a1.text)
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```

```python
#Lets Extract review

for Page2 in range(10):
    review=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for b1 in review:
        Review.append(b1.text)
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```

```python
#Lets Extract Fullreview

for Page3 in range(10):
    fdesc=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for c1 in fdesc:
        Fullreview.append(c1.text)
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```

```python
#Create the table using DataFrame

df9=pd.DataFrame({'Rating':Rating,'Review':Review,'Full Review':Fullreview})
df9
```
