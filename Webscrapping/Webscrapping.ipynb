```python
import requests

import pandas as pd

from bs4 import BeautifulSoup

URL3='https://presidentofindia.nic.in/former-presidents.htm'


Page4=requests.get(URL3)

Page4
```

```python
Soup4=BeautifulSoup(Page4.content)

Soup4.text
```

```python
Pname=[]

for p in Soup4.find_all('img',class_="alignLeft"):
    Pname.append(p.get('title'))

    
Pname
```

```python
len(Pname)
```

```python
Terms=[]

for te in Soup4.find_all('div',class_="presidentListing"):
    Terms.append(te.text.split('\n')[2])
    
    
Terms
```

```python
Termsofoffice=[]


for TOO in Terms:
    Termsofoffice.append(TOO.split(':')[1])
    
Termsofoffice
```

```python
df4=pd.DataFrame({'President Name':Pname,"Terms of Office":Termsofoffice})

df4
```

```python
#Solution3(C)----------------------------top10 ODI Bowlers
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL='https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'

Page10=requests.get(URL)

Page10
```

```python
Soup10=BeautifulSoup(Page10.content)

Soup10.text
```

```python
player=Soup10.find('div',class_="rankings-block__banner--name-large")

player.text
player=[player.text]
```

```python
Pteam=Soup10.find('div',class_="rankings-block__banner--nationality")
Pteam.text.split('\n')[2]

PTeam=[Pteam.text.split('\n')[2]]
PTeam
```

```python
Prate=Soup10.find('div',class_="rankings-block__banner--rating")
Prate.text

Prate=[Prate.text]
```

```python
Players=[]


for P in Soup10.find_all('td',class_="table-body__cell rankings-table__name name")[:9]:
    Players.append(P.text.split('\n')[1])
    
Players
```

```python
Team=[]

for Te in Soup10.find_all('span',class_="table-body__logo-text")[:9]:
    Team.append(Te.text)
    
Team
```

```python
Rate=[]

for Ra in Soup10.find_all('td',class_="table-body__cell rating")[:9]:
    Rate.append(Ra.text)
    
Rate
```

```python
Bowler=player+Players

len(Bowler)
```

```python
Teams=PTeam+Team
Teams
```

```python
Ratings=Prate+Rate

len(Ratings)
```

```python
df11=pd.DataFrame({'Bowlers':Bowler,'Team':Teams,'Ratings':Ratings})

df11
```
